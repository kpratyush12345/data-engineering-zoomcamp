## Question 1. Bruin Pipeline Structure

In a Bruin project, what are the required files/directories?

- bruin.yml and assets/
- .bruin.yml and pipeline.yml (assets can be anywhere)
- .bruin.yml and pipeline/ with pipeline.yml and assets/
- pipeline.yml and assets/ only


Answer - .bruin.yml and pipeline/ with pipeline.yml and assets/

Explanation:

A Bruin project has a specific structure that allows the Bruin CLI to understand how to connect to databases, how to run pipelines, and what tasks to execute.

The first required file is .bruin.yml. This file is located at the root of the project. It defines environments and database connections, such as DuckDB, BigQuery, or Snowflake. Bruin uses this file to know where to store and read data. Without this file, Bruin cannot run any pipeline because it has no connection information.

The second required component is the pipeline directory. This directory contains everything related to a specific pipeline.

Inside the pipeline directory, there must be a pipeline.yml file. This file defines the pipeline configuration, such as the pipeline name, schedule, default database connection, and variables. This tells Bruin how and when to run the pipeline.

The pipeline directory must also contain an assets folder. The assets folder contains the actual work of the pipeline. These assets can be Python files for ingestion, SQL files for transformations, or YAML seed files for lookup data. Each asset creates or updates tables in the database.

So in summary, Bruin requires:

- .bruin.yml → defines environments and database connections
- pipeline/ directory → contains pipeline configuration
- pipeline.yml → defines pipeline settings
- assets/ directory → contains ingestion, transformation, and reporting logic

These components together allow Bruin to validate, run, and orchestrate the pipeline correctly.
