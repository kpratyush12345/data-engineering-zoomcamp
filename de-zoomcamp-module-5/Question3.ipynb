## Question 3. Pipeline Variables

You have the following variable defined in pipeline.yml:

```python
variables:
  taxi_types:
    type: array
    items:
      type: string
    default: ["yellow", "green"]
```

How do you override this when running the pipeline to only process yellow taxis?

- bruin run --taxi-types yellow
- bruin run --var taxi_types=yellow
- bruin run --var 'taxi_types=["yellow"]'
- bruin run --set taxi_types=["yellow"]

Answer - bruin run –var ‘taxi_types=[“yellow”]’

Explanation

In Bruin, custom pipeline variables are overridden using the --var flag, and the value must match the variable’s defined type.

Our variable is defined as:
- Name: taxi_types
- Type: array of strings
- Default value: ["yellow", "green"]

Since this variable expects an array, we must pass the override value in array format, even if it contains only one element.

So the correct way is:

–var ‘taxi_types=[“yellow”]’

This tells Bruin to process only yellow taxis instead of both yellow and green.
