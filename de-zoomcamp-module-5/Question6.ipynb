## Question 6. Lineage and Dependencies

After building your pipeline, you want to visualize the dependency graph between assets. Which Bruin command should you use?

- bruin graph
- bruin dependencies
- bruin lineage
- bruin show

Answer: bruin lineage

Explanation

The bruin lineage command is used to visualize the dependency graph between assets in our pipeline.

It shows:
- Upstream dependencies → assets that run before the current asset
- Downstream dependencies → assets that run after the current asset
- The complete execution relationship between ingestion, staging, and reporting layers

This helps you understand execution order and data flow.

Example from our pipeline

When we ran:

bruin lineage pipeline/assets/staging/trips.sql

We saw:

Upstream Dependencies
- ingestion.trips
- ingestion.payment_lookup

Downstream Dependencies
- reports.trips_report

This confirms the flow:

ingestion → staging → reports
