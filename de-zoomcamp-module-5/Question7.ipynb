## Question 7. First-Time Run

You're running a Bruin pipeline for the first time on a new DuckDB database. What flag should you use to ensure tables are created from scratch?

- --create
- --init
- --full-refresh
- --truncate

Explanation -

When running a Bruin pipeline for the first time on a new database (or when we want to rebuild everything from scratch), we should use the --full-refresh flag.

This flag ensures that:
- Existing tables are dropped (if they exist)
- Tables are recreated from scratch
- Incremental logic is ignored
- We get a clean, complete rebuild of all assets

This is especially important when:
- Running the pipeline for the first time
- Changing schema or logic
- Fixing corrupted or duplicate data
- Resetting our environment
