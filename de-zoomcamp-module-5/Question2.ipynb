## Question 2. Materialization Strategies

You're building a pipeline that processes NYC taxi data organized by month based on pickup_datetime. Which incremental strategy is best for processing a specific interval period by deleting and inserting data for that time period?

- append - always add new rows
- replace - truncate and rebuild entirely
- time_interval - incremental based on a time column
- view - create a virtual table only

Answer - time_interval - incremental based on a time column

Explanation - 

The time_interval materialization strategy is specifically designed for pipelines that process data in defined time windows, such as daily, weekly, or monthly intervals based on a timestamp column like pickup_datetime.

In our NYC taxi pipeline, the data is naturally partitioned by time (for example, January 2022, February 2022, etc.). When we run the pipeline for a specific interval, we want Bruin to:

- Delete only the data for that specific time period
- Insert the newly processed data for that same time period
- Leave all other historical data untouched

This is exactly what time_interval does.

Why time_interval is the best choice

When using time_interval, Bruin:

- Uses a defined time column (like pickup_datetime) as the incremental key
- Deletes rows only within the specified start_date and end_date interval
- Inserts fresh, clean data for that interval
- Prevents duplication
- Allows safe reprocessing of specific months if needed

This is ideal for pipelines that run on schedules such as daily or monthly.
