## Question 4. Running with Dependencies

You've modified the ingestion/trips.py asset and want to run it plus all downstream assets. Which command should you use?

- bruin run ingestion.trips --all
- bruin run ingestion/trips.py --downstream
- bruin run pipeline/trips.py --recursive
- bruin run --select ingestion.trips+

Answer: bruin run ingestion/trips.py –downstream

Explanation -

In Bruin, if we want to run an asset and everything that depends on it (downstream assets), we use the --downstream flag.

We modified the asset:

ingestion/trips.py

This asset has downstream dependencies:
- staging.trips
- reports.trips_report

Using:

bruin run ingestion/trips.py –downstream

will run in this order:
1.	ingestion.trips
2.	staging.trips
3.	reports.trips_report

Bruin automatically determines the correct execution order based on lineage.

---

Why the other options are incorrect

bruin run ingestion.trips –all
Incorrect because --all is not a valid Bruin flag.

bruin run pipeline/trips.py –recursive
Incorrect because:
- Wrong file path
- --recursive is not a valid Bruin flag

bruin run –select ingestion.trips+
Incorrect because --select syntax belongs to dbt, not Bruin.
