## What can cost us money
From our Module 3 homework, we created mainly:

1. GCS bucket with parquet files
This costs storage money (very small, but still).

2. BigQuery dataset + tables

We created:
- External table
- Materialized (normal) table
- Partitioned / clustered table

BigQuery storage costs apply for:
- materialized table
- partitioned table

External table itself is free, but the data in GCS is not.

3. Queries

We only pay when:
- we run queries (scan data)

Right now your biggest cost risk is: storage, not compute.

### What you should delete (recommended after homework)

You can safely delete:
- the BigQuery dataset
- the GCS bucket used only for this homework

## Step 1 – Delete BigQuery dataset (this deletes all tables inside)

Run this in your terminal:

```python
bq rm -r -f zoomcamp_hw3
```

Explanation:
- -r → recursive (delete all tables inside)
- -f → no confirmation

This removes:
- yellow_tripdata_2024
- partitioned table
- external table
- everything in that dataset

## Step 2 – Delete the GCS bucket (and all parquet files)
Only do this if the bucket is ONLY for homework.

Run:

```python
gsutil -m rm -r gs://de-zoomcamp-485507-hw3-yellow
```

This deletes:
- all parquet files
- the bucket itself
