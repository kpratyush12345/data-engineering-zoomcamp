## Step 7 – Create the external table (PARQUET – important)

The homework says: You will need to use the PARQUET option when creating an external table.

So run:

```python
bq mk \
  --external_table_definition=PARQUET="gs://de-zoomcamp-485507-hw3-yellow/*.parquet" \
  zoomcamp_hw3.yellow_tripdata_2024_ext
```

## Step 8 – Create a native BigQuery table

This is what you will query for the homework.

In BigQuery UI or with bq query:

```python
CREATE OR REPLACE TABLE zoomcamp_hw3.yellow_tripdata_2024 AS
SELECT *
FROM zoomcamp_hw3.yellow_tripdata_2024_ext;
```
