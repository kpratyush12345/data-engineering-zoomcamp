# Step 1: Create / Select your GCP Project

Option A: If project already exists

Go to GCP Console → Project dropdown and copy the Project ID.

Option B: Create new project

GCP Console → New Project → Create → copy the Project ID

Example:
de-zoomcamp-485507

# Step 2: Create Service Account + Download Key (JSON)

## 2.1 Create Service Account

GCP Console → IAM & Admin → Service Accounts → Create service account

Name: terraform-runner

## 2.2 Give it permissions (roles)

Add these roles:

- Storage Admin (or Storage Object Admin + Storage Admin is safe for zoomcamp)
- BigQuery Admin
- Viewer (optional but useful)

## 2.3 Create Key JSON

Inside that service account:
Keys → Add Key → Create new key → JSON → Download

This will download something like:
de-zoomcamp-485507-80e91f64e7f2.json

# Step 3: Add key to your VS Code repo safely

Create a seperate directory for keys and rename the file to my-creds.json

# Step 4: Add .gitignore (VERY IMPORTANT)

You must NOT push creds to GitHub.

Create .gitignore in root of your repo (or inside terraform folder)

Paste:

```python
terraform/keys/
*.tfstate
*.tfstate.backup
.terraform/
.terraform.lock.hcl
```

This prevents secrets and terraform state from being committed.

# Step 5: Create main.tf and variables.tf

Added files under terraform directory

# Step 6: Create terraform.tfvars (best practice)

Instead of editing defaults in variables.tf, correct way is to use terraform.tfvars.

# Step 7: Run Terraform Commands (this creates bucket + dataset)

```python
# Go inside terraform folder
cd terraform

# Initialise Terraform
terraform init

# Check what will be created
terraform plan

# Apply and create resources
terraform apply
```

Type: Yes

After this, your bucket + dataset are created

# Step 8: Destroy resources

To avoid costs (bucket/dataset are cheap but still):

```python
terraform destroy
```
