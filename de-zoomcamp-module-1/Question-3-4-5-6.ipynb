# Question 4: Pickup day with longest trip distance (<100 miles)

Run:

```python
SELECT DATE(lpep_pickup_datetime) AS pickup_day,
       MAX(trip_distance) AS max_distance
FROM green_trips
WHERE lpep_pickup_datetime >= '2025-11-01'
  AND lpep_pickup_datetime < '2025-12-01'
  AND trip_distance < 100
GROUP BY DATE(lpep_pickup_datetime)
ORDER BY max_distance DESC
LIMIT 1;
```

Answer: 2025-11-14

# Question 5: Pickup zone with largest total_amount on Nov 18, 2025

We need:
- Filter pickup date = 2025-11-18
- Sum(total_amount)
- Group by pickup zone name

Pickup zone comes from PULocationID → zones table LocationID

Run:

```python
SELECT z."Zone" AS pickup_zone,
       SUM(g.total_amount) AS total_revenue
FROM green_trips g
JOIN zones z
  ON g."PULocationID" = z."LocationID"
WHERE g.lpep_pickup_datetime >= '2025-11-18'
  AND g.lpep_pickup_datetime < '2025-11-19'
GROUP BY z."Zone"
ORDER BY total_revenue DESC
LIMIT 1;
```

![Question 5](images/question5.png)

Answer: East Harlem North

# Question 6: From pickup zone “East Harlem North”, which dropoff zone had largest tip?

We need:
- Filter Nov 2025 pickups
- Only pickups in "East Harlem North"
- Group by dropoff zone, and take MAX(tip_amount)

Dropoff zone comes from DOLocationID → zones table.

Run:

```python
SELECT z_do."Zone" AS dropoff_zone,
       MAX(g.tip_amount) AS max_tip
FROM green_trips g
JOIN zones z_pu
  ON g."PULocationID" = z_pu."LocationID"
JOIN zones z_do
  ON g."DOLocationID" = z_do."LocationID"
WHERE g.lpep_pickup_datetime >= '2025-11-01'
  AND g.lpep_pickup_datetime < '2025-12-01'
  AND z_pu."Zone" = 'East Harlem North'
GROUP BY z_do."Zone"
ORDER BY max_tip DESC
LIMIT 1;
```

Answer: Yorkville West
